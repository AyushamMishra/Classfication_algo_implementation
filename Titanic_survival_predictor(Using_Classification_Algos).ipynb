## Titanic_Survival_Predictor

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df=sns.load_dataset("titanic")
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
categorical_cols = df.select_dtypes(include=['object', 'category']).columns
categorical_cols
```

```python
for i in categorical_cols:
  print(f"the value count of {i} : ")
  print(df[i].value_counts(),"/n")
```

```python
df['age'].fillna(df['age'].mean(), inplace=True)
df.dropna(subset=['embarked'],inplace=True)
```

```python
for i in categorical_cols:
  print(f"the value count of {i} : ")
  print(df[i].value_counts(),"/n")
```

```python
df.drop( ['class', 'who', 'deck', 'embark_town', 'alive'],axis=1,inplace=True)
```

```python
df_encode=pd.get_dummies(df,drop_first=True)
df_encode
```

```python
df_encode = df_encode.astype(int)
df_encode
```

```python
df['pclass'].value_counts()
```

```python
df_encode.info()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df_scaled=pd.DataFrame(scaler.fit_transform(df_encode),columns=df_encode.columns)
df_scaled
```

```python
X=df_scaled.drop('survived',axis=1)
y=df['survived']
```

## Logistic regression Model

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

## K-nearest Neighbours Model

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model=KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train,y_train)
```

```python
knn_pred=knn_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_score(y_test,knn_pred)
```

```python
print(classification_report(y_test,knn_pred))
```

## Naive Baye's Model

```python
from sklearn.naive_bayes import GaussianNB
model_nb=GaussianNB()
model_nb.fit(X_train,y_train)
```

```python
nb_pred=model_nb.predict(X_test)
```

```python
accuracy_score(y_test,nb_pred)
```

```python
print(classification_report(y_test,nb_pred))
```

## Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc_model=DecisionTreeClassifier(random_state=42)
dtc_model.fit(X_train,y_train)
```

```python
dtc_pred=dtc_model.predict(X_test)
dtc_pred
```

```python
accuracy_score(dtc_pred,y_test)
```

```python
print(classification_report(dtc_pred,y_test))
```
